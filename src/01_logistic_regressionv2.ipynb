## **19. Seleccion de Caracter√≠sticas**

M√©todo 1: Feature importance de RF

---



```python
# Entrenamiento de RF
model = RandomForestRegressor(
    random_state=42,
    n_jobs=-1
).fit(X_train_imp, y_train)
```

```python
# Importancia de caracter√≠sticas
importances = model.feature_importances_/model.feature_importances_.sum()*100

# Convertir a DataFrame
df_rf_imp = pd.DataFrame({'feature': X_train_imp.columns,'rf_importance': importances}).sort_values(by='rf_importance', ascending=False)

# Calculamos la importancia acumulada
df_rf_imp['rf_importance_acum'] = df_rf_imp['rf_importance'].cumsum()
```

```python
df_rf_imp
```

M√©todo 2: Permutation/Shuffle importance

```python
# Para esta t√©cnica y la de shap se necesita conjunto de validaci√≥n
X_train1, X_val, y_train1, y_val = train_test_split(
    X_train_imp, y_train, test_size=0.2, random_state=42
)

# Entrenamiento
model_xgb = XGBRegressor(
    random_state=42,
    objective="reg:squarederror",
    eval_metric="rmse"
)
model_xgb.fit(X_train1, y_train1)
```

```python
# Realizamos 10 permutaciones por cada caracter√≠stica
perm = permutation_importance(
    model_xgb,
    X_val,
    y_val,
    n_repeats=10,
    random_state=42,
    n_jobs=-1,
    scoring="neg_root_mean_squared_error"
)
```

```python
df_perm_imp = pd.DataFrame({
    "feature": X_val.columns,
    "perm_imp": perm.importances_mean
}).sort_values("perm_imp", ascending=False)

df_perm_imp
```

M√©todo 3: SHAP

```python
# Ajustamos el modelo
model_lgbm = lgb.LGBMRegressor(
    random_state=42,
    n_jobs=-1
).fit(X_train1, y_train1)
explainer = shap.Explainer(model_lgbm, X_val)   # usa el mismo X_val
shap_vals = explainer(X_val).values
```

```python
imp_shap = np.abs(shap_vals).mean(axis=0)
imp_shap_pct = imp_shap/imp_shap.sum()*100
df_shap_imp = pd.DataFrame({"feature": X_val.columns, "shap_imp": imp_shap_pct}).sort_values('shap_imp', ascending=False)
df_shap_imp
```

```python
# --- SHAP Summary Bar Plot ---
shap.summary_plot(shap_vals, X_val, plot_type="bar")
```

Uni√≥n 3 m√©todos

```python
df_importances = (
    df_rf_imp
    .merge(df_perm_imp, on='feature', how='outer')
    .merge(df_shap_imp, on='feature', how='outer')
).sort_values('rf_importance', ascending=False)
df_importances
```

```python
df_importances['perm_imp_pct'] = (
    df_importances['perm_imp'].clip(lower=0)
    / df_importances['perm_imp'].clip(lower=0).sum()
    * 100
)
df_filt = df_importances[
    (df_importances['rf_importance'] >= 1) &
    (df_importances['perm_imp_pct'] >= 1) &
    (df_importances['shap_imp'] >= 1)
]
df_filt
```

## **25.Busqueda de Hiperparametros** Nos quedamos con la Busqueda Intensiva. Tiene ROC AUC mayor.

B√∫squeda extensiva

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']  # compatible con l1 y l2
}
logreg = LogisticRegression(max_iter=1000)

grid = GridSearchCV(
    estimator=logreg,
    param_grid=param_grid,
    scoring='roc_auc',      # recomendado para clasificaci√≥n
    cv=5,
    n_jobs=-1
)

grid.fit(X_train_scaled, y_train)
best_model = grid.best_estimator_
print(grid.best_params_)
print(grid.best_score_)
```

Busqueda Intensiva

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform
param_dist = {
    'C': loguniform(1e-4, 1e2),   # b√∫squeda continua
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}

logreg = LogisticRegression(max_iter=1000)

random_search = RandomizedSearchCV(
    estimator=logreg,
    param_distributions=param_dist,
    n_iter=30,           # n√∫mero de combinaciones probadas
    scoring='roc_auc',
    cv=5,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train_scaled, y_train)
best_model = random_search.best_estimator_

print(random_search.best_params_)
print(random_search.best_score_)
```
